***Representación de funciones de dos variables***

Se selecciona la función de Rosenbrock para representar. A continuación se muestra al ecuación:

```python
Math(rf'f(x1,x2)=100(x2−x1^2)^2+(1−x1)^2 , xi∈[−2.048,2.048], i=1,2')
```

```python
b = 100;
r = lambda x,y: (x-1)**2 + b*(y-x**2)**2;
```

```python
# Definir figura
figRos = plt.figure(figsize=(10, 10))
axRos = figRos.gca(projection='3d')

# Evaluate function
X = np.arange(-2.048, 2.048, 0.1)
Y = np.arange(1, 3, 0.1)
X, Y = np.meshgrid(X, Y)
Z = r(X,Y)

# Plot the surface
surf = axRos.plot_surface(X, Y, Z, cmap=cm.gist_heat_r,
                       linewidth=0, antialiased=False)
axRos.set_zlim(0, 2000)
figRos.colorbar(surf, shrink=0.5, aspect=11)
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(10, 10))
CS = ax.contour(X, Y, Z,10)
ax.clabel(CS, inline=True, fontsize=10)
ax.set_title('Gráfico Curvas de Nivel - Función Rosenbrock')
```
